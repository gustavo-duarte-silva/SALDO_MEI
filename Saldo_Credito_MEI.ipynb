```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/CIENTISTA DE DADOS/PROJETOS/SALDO CREDITO MEI/bcdata.csv', sep=';')
```

```python
#Tratamento dos Dados
def pre_processamento(df):
  dff = df.copy()
  dff['data'] = pd.to_datetime(dff.data, dayfirst=True, format='%d/%m/%Y')
  dff['valor'] = dff.valor.str.replace(',', '.').astype('float')
  return dff
```

```python
dff = pre_processamento(df)
```

```python
dff.dtypes
```

```python
plt.figure(figsize=[15,8])
sns.set_theme(style="darkgrid")
plt.title("Saldo de Cr√©dito do MEI 2016-2021")
sns.lineplot(data=dff, x='data', y='valor')
plt.show()
```
